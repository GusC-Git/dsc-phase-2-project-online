## Q2. What is the Relationship between Grade and Housing Price?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/king_county_data_clean.csv").drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
sns.displot(df['grade'])
plt.show()
```

```python
sns.scatterplot(data=df, x='grade', y='price', alpha=0.3)
plt.show()
```

```python
import matplotlib.lines as mlines
```

```python
plt.figure(figsize=(15,8))
sns.jointplot(x='grade', y='price', data=df, kind='hist', color="#038fe7")
sns.lineplot(x='grade', y='price', data=df, color='salmon')
plt.xlabel(xlabel='Grade', size=15)
plt.ylabel(ylabel='Price Sold', size=15)
plt.xticks(fontsize=13)

# Creates legend
top_line = mlines.Line2D([], [], color='salmon',markersize=15, label='Mean Grade vs. Price')
plt.legend(handles=[top_line], loc='upper left')

plt.show()
```

```python
np.corrcoef(x= df['grade'], y= df['price'])[0][1]
```

### We see a general trend that with an increase in house grade comes an increase in price. We also notice that there are a lage concentration of houses that recieve a grade of 7. When looking at the correlation coefficient we see that there is a fairly stong correlation between grade and price.
